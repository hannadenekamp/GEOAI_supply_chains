```python
box_dct = {}
for output_pth in test_images:
    with open(output_pth,'rb') as f:
        outputs = pickle.load(f)
    name = output_pth.split('\\')[-1].split('_o')[0]
    boxes = outputs[0]['boxes'].data.numpy()
    scores = outputs[0]['scores'].data.numpy()
    # Filter out boxes according to `detection_threshold`
    boxes = boxes[scores >= 0.9].astype(np.int32)
    if boxes.size == 4:
        box_dct[name] = boxes[0]
    if boxes.size > 4:
        boxes = boxes[scores.argmax()]
        box_dct[name] = boxes

```
